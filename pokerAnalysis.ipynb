**Step 2: Analyze data and probability outcomes.**

```python
import pandas as pd

pocketCardFrequency = pd.read_csv("pocketCardsFrequency.csv")
```

```python
pocketCardFrequency.tail()
```

```python
pocketCardFrequency = pocketCardFrequency.drop(pocketCardFrequency.columns[0], axis = 1)
pocketCardFrequency
```

```python
pocketCardFrequency.dtypes
```

```python
pocketCardFrequency["Pair"] = pocketCardFrequency["Pair"].astype(int)
pocketCardFrequency["Suited"] = pocketCardFrequency["Suited"].astype(int)
pocketCardFrequency["Connected"] = pocketCardFrequency["Connected"].astype(int)
```

```python
pocketCardFrequency.columns
```

```python
pocketCardFrequency.shape
```

```python
totalSimulations = pocketCardFrequency.iloc[:, 1:9].sum().sum()
totalSimulations
```

```python
pair = pocketCardFrequency.groupby(["Pair"]).sum()
pair = pair.drop(columns = ["Pocket Cards", "Suited", "Connected"]).T
pair = pair.rename(columns = {"Pair": "Number of Players", 0: "No Pair", 1: "Pair"})

suited = pocketCardFrequency.groupby(["Suited"]).sum()
suited = suited.drop(columns = ["Pocket Cards", "Pair", "Connected"]).T
suited = suited.rename(columns = {0: "Not Suited", 1: "Suited"})

connected = pocketCardFrequency.groupby(["Connected"]).sum()
connected = connected.drop(columns = ["Pocket Cards", "Pair", "Suited"]).T
connected = connected.rename(columns = {0: "Not Connected", 1: "Connected"})
```

```python
trueValues = pd.concat([pair, suited, connected], axis = 1).T
trueValues
```

```python
trueValueProbabilities = trueValues.copy()

# Each game generated 10k hands per player, meaning a 2player game generated 20k, a 3player game generated 30k, ..., etc.
# We then mulitply by a 100 to get the percentage

def divideByPlayerCount(x, colName):
  if colName == "Frequency with 2players":
    return (x / 20000) * 100
  elif colName == "Frequency with 3players":
    return (x / 30000) * 100
  elif colName == "Frequency with 4players":
    return (x / 40000) * 100
  elif colName == "Frequency with 5players":
    return (x / 50000) * 100
  elif colName == "Frequency with 6players":
    return (x / 60000) * 100
  elif colName == "Frequency with 7players":
    return (x / 70000) * 100
  elif colName == "Frequency with 8players":
    return (x / 80000) * 100
  elif colName == "Frequency with 9players":
    return (x / 90000) * 100

for column in trueValueProbabilities.columns:
  trueValueProbabilities[column] = trueValueProbabilities[column].apply(divideByPlayerCount, colName = column)

trueValueProbabilities
```
